````python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])
plt.subplot(121);plt.imshow(img) # expects distorted color
# plt.axis('off') 
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.subplot(122);plt.imshow(img2) # expect true color
plt.axis('off')
plt.show()
````

## knn算法 

### 原理 
就是找与待分类目标距离最接近的K个目标的类别按权值统计，来得到待分类目标的类别

好处的是不需要训练，坏处是每次计算比较费时和麻烦 

### opencv中的knn 



````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

# 这样生成几个二维坐标

traindata = np.random.randint(0,100,(25,2)).astype(np.float32) # 生成两个由10个0.5-1.5之间的数构成的数组
responses = np.random.randint(0,2,(25,1)).astype(np.float32)
#print 'traindata:'
#print traindata

red = traindata[responses.ravel()==0]
plt.scatter(red[:,0],red[:,1],80,'r','^')

blue = traindata[responses.ravel()==1]
plt.scatter(blue[:,0],blue[:,1],80,'b','s')

plt.show()
````

接下来使用traindata和responses来训练knn算法，只是会建立一个二叉搜索树。

用新来的点来预测它的分类，用knn算法，会返回四个参数：
1. ret,bool类型表示程序执行是否正确
2. label,给新来的贴的标签 
3. 找到k个邻居的label
4. 找到的与k个邻居的距离


注意opencv3访问ml模块的方式变了[opencv3的访问方式](https://github.com/goncalopp/simple-ocr-opencv/issues/9)

````python

testdata = np.random.randint(0,100,(5,2)).astype(np.float32)

# Labels each one either Red or Blue with numbers 0 and 1
responses = np.random.randint(0,2,(25,1))

knn = cv2.ml.KNearest_create()
knn.train(traindata,0,responses) # 第二个参数0表示行样本 1表示列样本

ret,result,neighbours,dists = knn.findNearest(testdata,4)
redNew = testdata[result.ravel()==0]
blueNew = testdata[result.ravel()==1]


plt.scatter(red[:,0],red[:,1],80,'r','^')
plt.scatter(blue[:,0],blue[:,1],80,'b','s')


plt.scatter(redNew[:,0],redNew[:,1],80,'g','^') # 菱形
plt.scatter(blueNew[:,0],blueNew[:,1],80,'c','s') # 倒三角
plt.show()
````
