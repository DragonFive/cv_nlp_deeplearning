## knn算法 

### 原理 
就是找与待分类目标距离最接近的K个目标的类别按权值统计，来得到待分类目标的类别

好处的是不需要训练，坏处是每次计算比较费时和麻烦 

### opencv中的knn 



````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

# 这样生成几个二维坐标

traindata = np.random.randint(0,100,(25,2)).astype(np.float32) # 生成两个由10个0.5-1.5之间的数构成的数组
responses = np.random.randint(0,2,(25,1)).astype(np.float32)
#print 'traindata:'
#print traindata

red = traindata[responses.ravel()==0]
plt.scatter(red[:,0],red[:,1],80,'r','^')

blue = traindata[responses.ravel()==1]
plt.scatter(blue[:,0],blue[:,1],80,'b','s')

plt.show()
````

接下来使用traindata和responses来训练knn算法，只是会建立一个二叉搜索树。

用新来的点来预测它的分类，用knn算法，会返回四个参数：
1. ret,bool类型表示程序执行是否正确
2. label,给新来的贴的标签 
3. 找到k个邻居的label
4. 找到的与k个邻居的距离


注意opencv3访问ml模块的方式变了[opencv3的访问方式](https://github.com/goncalopp/simple-ocr-opencv/issues/9)

````python

testdata = np.random.randint(0,100,(5,2)).astype(np.float32)

# Labels each one either Red or Blue with numbers 0 and 1
responses = np.random.randint(0,2,(25,1))

knn = cv2.ml.KNearest_create()
knn.train(traindata,0,responses) # 第二个参数0表示行样本 1表示列样本

ret,result,neighbours,dists = knn.findNearest(testdata,4)
redNew = testdata[result.ravel()==0]
blueNew = testdata[result.ravel()==1]


plt.scatter(red[:,0],red[:,1],80,'r','^')
plt.scatter(blue[:,0],blue[:,1],80,'b','s')


plt.scatter(redNew[:,0],redNew[:,1],80,'g','^') # 菱形
plt.scatter(blueNew[:,0],blueNew[:,1],80,'c','s') # 倒三角
plt.show()
````

### knn进行手写体识别——opencv自带数据 

使用opencv自带的样本,opencv自带5000个样本，10个数字每个数字500个手写体

MNIST data 也可以作为训练集合



````python
import numpy as np
import cv2
from sklearn.metrics import classification_report

# 读取数据 
img  = cv2.imread('digits.png')
#gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# 分解数据 
cells = [np.hsplit(row,100) for row in np.vsplit(gray,50)]
x = np.array(cells)
print x.shape  # 发现这里的像素是20×20的


# 分开训练样本和测试样本 reshape的某个参数为-1时需要程序自己推算
train = x[:,:50].reshape(-1,400).astype(np.float32)
test  = x[:,50:100].reshape(-1,400).astype(np.float32)

# 产生label
k = np.arange(10)
train_labels = np.repeat(k,250)[:,np.newaxis]
# 上面的等效于train_labels = np.repeat(k,10).reshape(-1,1)
test_labels = train_labels.copy() # 深层复制


# 进行knn训练
knn = cv2.ml.KNearest_create()
knn.train(train,0,train_labels)


# 进行测试和计算精确度
ret,results,neighbors,distances = knn.findNearest(test,5)
matches = results == test_labels


correct = np.count_nonzero(matches) # 计算非0元素个数
print correct*100.0/train.shape[0]
print classification_report(test_labels,results)

````

#### 输入数据 

对于图像要在模型中使用，需要进行适当的转换。因为图像是像素值组成的数组。彩色图像有多通道（RGB），但MNIST图像是灰度图只有单通道，很方便操作，所以我们可以把每个图片映射到一个numpy的数组。

#### 模型 

最简单的模型是Knn算法，它甚至都不需要训练，因为没有参数需要训练。但是这个算法比较耗时，因为要计算待分类图片与数据库中每个图片的距离，如果数据量大就会很慢。这是很我们可以有两种改进方法。

1. 使用近似最近邻
2. 使用带参数的模型
